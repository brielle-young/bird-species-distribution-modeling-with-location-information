```python
import os
import json
import pandas as pd
```

```python
targets_dir = os.path.join("USA_summer", "targets")
targets = pd.DataFrame([])

for target_file in os.listdir(targets_dir):
        with open(os.path.join(targets_dir, target_file)) as f:
                d = json.load(f)
                target = {"hotspot_id": [d["hotspot_id"]], "num_complete_checklists": [d["num_complete_checklists"]]}
                for i in range(len(d["probs"])):
                        target[f"prob_{i}"] = [d["probs"][i]]
                targets = pd.concat([targets, pd.DataFrame(target)])
```
